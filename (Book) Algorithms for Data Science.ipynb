```python
# Build a candidate committee dictionary from the Candidate Master file
canDict = {}
path = '/Users/Andrew/Desktop/Coursera/data/cn.txt'
n = 0
nn = 0
with open(path) as f:
    for line in f:
        data = line.split("|")
        canDict[data[9]] = data[2]
canDict
#print(canDict.get('C00510628'))
#print(canDict['C00510628'])
```

```python
# Build an other committees dictionary from the Committee Master file
otherDict = {}
path = '/Users/Andrew/Desktop/Coursera/data/cm.txt'
with open(path) as f:
    for line in f:
        data = line.split('|')
        otherDict[data[0]] = data[10]
otherDict
```

```python
# Build the dictionary employerDict. Keys will be employers and values will be a list of pairs. 
# Each pair in a list will be a political party and a contribution amount
employerDict = {}
path = '/Users/Andrew/Desktop/Coursera/data/indiv14/itcont.txt'
n = 0
with open(path) as f:
    for line in f:
        data = line.split('|')
        party = canDict.get(data[0])
        if party == None:
            party = otherDict[data[0]]
        if party != None:
            x = (party, int(data[14])) 
            employer = data[11]
            if employer != '':
                value = employerDict.get(employer)
                if value is None:
                    employerDict[employer] = [x]
                else:
                    employerDict[employer].append(x)
employerDict
employerDict['FERNLEAF INVESTMENT COMPANY, INC.']   
```

```python
# The last substantive set of code will reduce employerDict to
# reducedDict, as a dictionary of dictionaries. 
# The keys of reducedDict are to be employers and the 
# values are to be dictionaries. The keys of these 
# internal or sub-dictionaries are the party labels 
# Other, DEM, and REP and the values are the 
# total of contributions to each party made by 
# employees of a company

reducedDict = {}
for key in employerDict:
    totals = {'REP':0,'DEM':0,'Other':0}
    for value in employerDict[key]:
        try:
            totals[value[0]] += value[1]
        except KeyError:
            totals['Other'] += value[1]
    reducedDict[key] = totals
reducedDict
```

```python
# We will want to sort the entries in the employer dictionary 
# according to the total of all contributions made by employees 
# of an employer. Im- mediately after the statement 
# reducedDict = {}, initialize a dictionary named sumDict to 
# contain the total of all contributions made by employ- ees of 
# each employer. Add an instruction that computes the sum of the 
# three dictionary values and stores it with the employer key. 
reducedDict = {}
sumDict = {}
for key in employerDict:
    totals = {'REP':0,'DEM':0,'Other':0}
    for value in employerDict[key]:
        try:
            totals[value[0]] += value[1]
        except KeyError:
            totals['Other'] += value[1]
    reducedDict[key] = totals
    sumDict[key] = totals['REP'] + totals['DEM'] + totals['Other']
    #if sumDict[key] > 10000 : print(key, totals)
sumDict
```

```python
sumDict.items()[1] # dictionary to array
print(type(sumDict.items()))
```

```python
# Now that sumDict has been built, we will create a list from the 
# dictionary in which the largest contribution sums are the first 
# elements
sortedList = sorted(sumDict.items(), key=operator.itemgetter(1))
n = len(sortedList)
print(sortedList[n-200])
print(sortedList[1:3])
```

```python
print(sortedList[n-1][0])
print(sortedList[n-1][1])
print(type([sortedList[n-1][0]]))
print(type(sortedList))
```

```python
# Write a short list of the largest 200 employers to a text file
path = '/Users/Andrew/Desktop/Coursera/data/employerMoney.txt'
with open(path,'w') as f:
    for i in range(n-200,n):
        employerName = sortedList[i][0].replace("'","")
        totals = reducedDict[employerName]
        outputRecord = [employerName] + [str(x) for x in totals.values() +
                                        [str(sortedList[i][1])]]
        string = ';'.join(outputRecord) + '\n'
        f.write(string)
        
```
