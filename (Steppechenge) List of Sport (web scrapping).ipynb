```python
import requests

#url = 'https://www.topendsports.com/sport/list/index.htm'
url = 'https://www.topendsports.com/sport/recognized-sports.htm'
web_page = requests.get(url)

print(web_page.text)
#print(web_page.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(web_page.content,'html.parser')
```

Note: You’ll want to pass page.content instead of page.text to avoid problems with character encoding. The .content attribute holds raw bytes, which can be decoded better than the text representation you printed earlier using the .text attribute.

```python
import pandas as pd 

tdr_list = soup.select('tr td:nth-of-type(1)')
#print(tr_list[1])
#print(tr_list[1].find_next_siblings('td'))
sports_arr = []
for tr in tr_list:
    text = tr.text
    sports_arr.append(text)
#sports_arr  

df = pd.DataFrame({'sports': sports_arr})
df['sports'] = df['sports'].str.lower().str.replace(',',' ').unique()
print(df.iloc[2])
df.to_csv('/Users/Andrew/Desktop/sports.csv',index=False)
```

```python
#td_list_1 = soup.select('tr td:nth-of-type(1)')
#td_list_2 = soup.select('tr td:nth-of-type(2)')
tr_list = soup.select('tr')
#print(tr_list[1])
#print(tr_list[1].find_next_siblings('td'))
sports_arr = []
for tr in tr_list[1:]:
    t1 = tr.select('td:nth-of-type(1)')[0].text
    t2 = tr.select('td:nth-of-type(2)')[0].text
    #print(t1,":",t2)
    t = {'sport':t1,'description':t2}
    #text = tr.text
    sports_arr.append(t)
#sports_arr  

df = pd.DataFrame(sports_arr)
df
df['sport'] = df['sport'].str.lower().str.replace(',',' ')
print(df.iloc[2])
df.to_csv('/Users/Andrew/Desktop/sports.csv',index=False)
```
