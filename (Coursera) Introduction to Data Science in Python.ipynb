# Group by

````python
import pandas as pd
import numpy as np
df = pd.read_csv('data/census.csv')
df = df.loc[df['SUMLEV'] == 50, :]
df.head()
````

````python
%%timeit -n 10
for state in df['STNAME'].unique():
    #avg = np.average(df.where(df['STNAME'] == state).dropna()['CENSUS2010POP'])
    avg = np.average(df.loc[df['STNAME'] == state, 'CENSUS2010POP'].dropna())
    print('Counties in state {0} have an average population of {1}'.format(state, avg))
````

````python
%%timeit -n 10
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state {0} have an average population of {1}'.format(group, avg))
````

### Group by function

````python
# we can group based on function. Below we group based on first letter of state name
# for this we create function which returns 0,1,2 based on first letter of state.
# it's important to know that to use this hack we have to set target column as index
# this hack is used to distribute work between workers, or disks on a database, nodes on supercomputer, etc

#df.head()
df = df.set_index('STNAME')
df.head()

def fun(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are {0} records in group {1} for processing.'.format(len(frame), group))
````

What is the chance of tornado in two adjasent days


````python
chance_of_tornado = 0.01
tornado_events = np.random.binomial(1, chance_of_tornado, 1000000)
two_days_in_a_row = 0
for j in range(1, len(tornado_events) - 1):
    if tornado_events[j] == 1 and tornado_events[j-1] == 1:
        two_days_in_a_row += 1
        
print('{} tornadoes back to back in {} years'.format(two_days_in_a_row, 1000000/365))
````
