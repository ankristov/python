### Conversion rate by marketing channel (CR from conversion to retention)

```python
# Subset to include only House Ads
house_ads = marketing[marketing['subscribing_channel'] == 'House Ads']
retained = house_ads[house_ads['is_retained'] == True]['user_id'].nunique()
subscribers = house_ads[house_ads['converted'] == True]['user_id'].nunique()
retention_rate = retained / subscribers
print(round(retention_rate * 100, 2),"%")
```

```python
# Calculate for all channels at once
retained = marketing[marketing['is_retained'] == True].groupby(by='subscribing_channel')['user_id'].nunique().to_frame()
#retained = marketing[marketing['is_retained'] == True].groupby(by='subscribing_channel').agg({'user_id':['count',pd.Series.nunique]})
print('---retained\n',retained)

# Calculate subscribers by channel
subscribers = marketing[marketing['converted'] == True].groupby(by='subscribing_channel')['user_id'].nunique().to_frame()
#subscribers = marketing[marketing['converted'] == True].groupby(by='subscribing_channel').agg({'user_id':['count',pd.Series.nunique]})
print('---subscribers\n',subscribers)

# AK: calculate adv impressions by channel
impressions = marketing.groupby(by='subscribing_channel')['user_id'].count().to_frame()
print('--impressions\n',impressions)

# CR subscription -> retention
channel_retention_rate = (retained / subscribers) * 100
print('---channel_retention_rate\n',channel_retention_rate)
# AK: the same in one dataframe
merge_1 = subscribers.merge(right=retained,how='inner',left_index=True,right_index=True)
merge_1.rename({'user_id_y':'retained','user_id_x':'subscribers'},inplace=True,axis='columns')
merge_1['cr'] = 100.0 * merge_1['retained'] / merge_1['subscribers']
print('--CR subscribers to retained\n',merge_1)

# AK: CR impressions -> subscriptions
merge_2 = subscribers.merge(right=impressions,how='inner',left_index=True,right_index=True)
merge_2.rename({'user_id_y':'impressions','user_id_x':'subscribers'},inplace=True,axis='columns')
merge_2['cr'] = merge_2['subscribers'] / merge_2['impressions']
print('--CR impressions to subscriptions\n',merge_2)
```

```python
# AK: count all users, unique, duplicates in each segment marketing_channel & age_group
# Creating Custom Aggregations to Use with Pandas groupby
# https://towardsdatascience.com/creating-custom-aggregations-to-use-with-pandas-groupby-e3f5ef8cb43e

# create custom aggregation function
def f(x):
    return x.count() - x.nunique()

def n_duplicates(x):
    #x = pd.Series(x)
    return x.duplicated().sum()

channel_age = marketing.groupby(by=['marketing_channel','age_group']).agg({'user_id': ['count',pd.Series.nunique, f, n_duplicates]})
# so not clear why in the lab she counts not unique users...

print(channel_age)
print(channel_age['user_id']['n_duplicates'].sum())
print(marketing['user_id'].duplicated().sum())
print(n_duplicates(marketing['user_id']))
```

### Building a conversion function

You've been doing a lot of repetitive calculations. Anytime you notice repetition in your work, consider automation. The more of the low-variance work you can automate, the more time you will have to explore new and interesting data science topics at work. This will both impress your marketing stakeholders and be more fun!

Since you know the format of the marketing DataFrame will remain the same over time, you can build a function to enable you to calculate conversion rate across any sub-segment you want on the fly.

In this exercise, you will build a function that takes a DataFrame and list of column names and outputs the conversion rate across the column(s).

Instructions
- Isolate rows in the user inputted dataframe where users were converted, then group by the list of user inputted column_names and count the number of unique converted users.
- Group the user inputted dataframe by the list of user inputted column_names and calculate the total number of users.
- Fill any missing values in conversion_rate with 0.

```python
def conversion_rate(dataframe, column_names):
    # Total number of converted users
    column_conv = dataframe[dataframe['converted'] == True].groupby(column_names)['user_id'].nunique()

    # Total number users
    column_total = dataframe.groupby(by=column_names)['user_id'].nunique()
    
    # Conversion rate 
    conversion_rate = column_conv/column_total
    
    # Fill missing values with 0
    conversion_rate = conversion_rate.fillna(0)
    return conversion_rate
```

## Creating a lift function

Lift can be calculated by calculating the difference between the treatment effect (or the mean) of the treatment compared to the treatment effect of the control divided by the treatment effect of the control. The formula for lift can be found below:

 
(Treatment conversion rate - Control conversion rate) / Control conversion rate

The result is the percent difference between the control and treatment.

In this exercise, you will create a function to automate the process of calculating lift. Many marketing teams run tests constantly. The more that you can automate the parts of the process that occur within every test, the more time you will have to do more interesting analyses.

Instructions
- Calculate the mean of a and b using np.mean().
- Use a_mean and b_mean to calculate the lift of the treatment.
- Print the results of the lift() function you created using the control and personalization variables.

```python
def lift(control, test):
    
    # Calcuate the mean of a and b
    mean_control = np.mean(control)
    mean_test = np.mean(test)
    
    # Calculate the lift using a_mean and b_mean
    lift = (mean_test - mean_control) / mean_control
    return str(round(lift*100,2)) + '%'

# Print lift() with control and personalization as inputs
print(lift(control, personalization))
```

Great work! As you can see, there's a large lift, but are your results statistically significant? You will find out in the next exercise.

## Building an A/B test segmenting function

In the previous lesson, you observed that your personalization experiment is highly statistically significant. However, when running experiments, it is important to check how new features are affecting specific demographics. Sometimes features that are highly appealing to one group are less appealing to others.

Since you want to segment our data multiple times, you will build a function ab_segmentation() that analyzes the impact of your A/B tests on segments of data that you can reuse each time you want to conduct this kind of analysis.

Your function will take in a column name and run through each unique value in that column calculating lift and statistical significance.

Instructions
- Build a for loop within our ab_segmentation() function that runs through each unique value in the user-inputted column name segment.
- Isolate the rows in marketing where the marketing channel is 'Email' and the user-inputted column segment equals subsegment.
- Print the results of the lift() and stats.ttest_ind() functions.

```python
marketing.head()
```

```python
def ab_segmentation(segment):
    # Build a for loop for each subsegment in marketing
    for s in np.unique(marketing[segment]):
        print(s)
        
        # Limit marketing to email and subsegment
        email_subsegment = marketing[(marketing['marketing_channel'] == 'Email') & (marketing[segment] == s)]
        subscribers = email_subsegment.groupby(by=['user_id','variant'])['converted'].max()
        subscribers = pd.DataFrame(subscribers.unstack(level=1))
        #print(subscribers.head())
        control = subscribers['control'].dropna()
        personalization = subscribers['personalization'].dropna()
        
        print('lift:', lift(control,personalization))
        print('t-statistics: ', stats.ttest_ind(control,personalization),'\n\n')
    return
        
```

```python
# AK
def ab_segmentation_1(segment):
    '''
        segment - column name with categorical values for calculation AB test results in sub-segments
    '''
    for s in marketing[segment].unique():
        print(f'---{s}\n',s)
        email_subsegment = marketing[ (marketing['marketing_channel'] == 'Email') & (marketing[segment] == s) ]
        email_subsegment_converted = email_subsegment.groupby(by=['user_id','variant']).agg({'converted':'max'}).reset_index()
        email_subsegment_ab = email_subsegment_converted.groupby('variant').agg({'converted':['count','sum','mean']})
        print(email_subsegment_ab.head())
        #print('lift: ', email_subsegment_ab['personalization',('converted','mean')])
        
        # lift
        test_mean = email_subsegment_ab.loc['personalization',('converted','mean')]
        cntl_mean = email_subsegment_ab.loc['control',('converted','mean')]
        lift = 100 * (test_mean - cntl_mean) / cntl_mean
        print('lift: ', round(lift,2),'%')
        

ab_segmentation_1('language_displayed')
```
