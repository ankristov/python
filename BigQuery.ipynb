### 2. unpacking entities by iterating


```python
# extract elements of enetities arrays of dict (not really useful because makes the table longer: some entites have several elements in array)
def array_of_dict_to_df(df,column):
    rows = []
    for i,row in df[column].iteritems():
        for item in row:
            rows.append(item)
    return pd.DataFrame(rows)
array_of_dict_to_df(df,'entities')[:5]
```

```python
# will work for us: extract name field and concatenate all names into one string
def array_of_dict_to_concat_field(df,column):
    rows = []
    for i,row in df[column].iteritems():
        s = ''
        for item in row:
            j = json.loads(item)
            #print(j['name'])
            #s = ','.join([s,j['name']])
            s = s + ', ' + j['name']
        rows.append(s.strip(', '))
    return rows

df['entities_name_concat'] = array_of_dict_to_concat_field(df,'entities')
df['entities_len'] = df['entities'].apply(lambda x: len(x)) # len of array in entites
df[df['entities_len'] > 1].head() # check the result where entites name length is more than 1
```
